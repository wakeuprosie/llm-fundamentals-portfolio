```python
import tiktoken
import pandas as pd

#initiate Chat GPT-4's tokenizer 'tiktoken'
enc = tiktoken.get_encoding("cl100k_base")

#varied query input examples to demonstrate differences in tokenization
examples = {
    "Plain english": "I’m a regular person trying to understand tokenization thanks to Andrej Karpathy’s awesome resources.",
    "Noisy user text": "heyyyy did you get tix for the BTS concert??! 🤩",
    "JSON": '{"product":"popcorn","calories":90,"isLowCalorie":true}',
    "Code": 'def calculate_total():',
    "Multilingual text": "是时候学中文了！你好。"
}

#display tokenized text examples in a table
table = []

for type, text in examples.items():
  token_ids = enc.encode(text)
  table.append(
      {
          "input_type": type,
          "input_text": text,
          "num_tokens": len(token_ids),
          "tokens": token_ids
      }
  )

dt = pd.DataFrame(table)

dt.sort_values("num_tokens", ascending=False, inplace=True)

pd.set_option('display.max_colwidth', None)

dt
```
